```python
"""
Author: Kostas Spyropoulos
Dr. Sam
CIS 2532 NET 01 
17.2 Program: Creates a tabular display from using the description attribute and fetchall method from cursor
05/10/2022
First Run
"""
import sqlite3
import time
import pandas as pd

# connects to books.db
connection = sqlite3.connect('C:\\Users\\kspyr\\Desktop\\CIS 2532\\HW 12\\ch17\\books.db')

# cursor object gets information from connection
cursor = connection.cursor()

# cursor objects gathers information from titles table query
cursor = cursor.execute("""SELECT * FROM titles""")

# list of column names such as isbn, title, edition, and copyright
columnNames = []

# columnNamesTuple contains the tuple of tuples from the description cursor attribute
columnNamesTuple = cursor.description

# fetchAll is equal to the fetchall method of the cursor object
fetchAll = cursor.fetchall()

# for loop that appends each column name to columnNames list from the columnNamesTuple
for columnNameTuple in columnNamesTuple:
    for columnName in columnNameTuple:
        if columnName != None:
            columnNames.append(columnName)

# rows representing the number of fetchAll data containing query result's data for isbn, title, edition, and copyright
rows = len(fetchAll)

# columns representing the number of column names isbn, title, edition, and copyright
columns = len(columnNames)

j = 0

# while loop prints the column names in tabular format
while j < columns:
    print(columnNames[j].ljust(32, ' '), end=' ')
    j+=1

# separates the column names from query result's data
print('---------------------------------------------------------------------------------------------------------------------------')
 
j = 0

# while loop prints the query result's data for each column
while j < rows:
    for part in fetchAll[j]:
        print(str(part).ljust(32,' '), end=' ')
    j+=1
```
