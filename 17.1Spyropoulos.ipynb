```python
"""
Author: Kostas Spyropoulos
Dr. Sam
CIS 2532 NET 01 
17.1 Program: Explores SQL with a book relational database
05/10/2022
First Run
"""
import sqlite3
import time
import pandas as pd

# variable connects to the books.db
connection = sqlite3.connect('C:\\Users\\kspyr\\Desktop\\CIS 2532\\HW 12\\books.db')

# print statement displays authors' last names in descending order
print(pd.read_sql('SELECT last FROM authors ORDER BY last DESC', connection))

print()

# print statement displays titles of books in ascending order
print(pd.read_sql('SELECT title FROM titles ORDER BY title ASC', connection))

print()

# print statement displays book information and identifies the author or authors of each book
print(pd.read_sql("""SELECT title, titles.isbn, copyright, authors.first, authors.last FROM titles INNER JOIN author_ISBN ON titles.isbn = author_ISBN.isbn INNER JOIN authors ON authors.id=author_ISBN.id ORDER BY title ASC""", connection))

# cursor object made
cursor = connection.cursor()

# inserts a new author 
cursor = cursor.execute("""INSERT INTO authors (id, first, last) VALUES ('6', 'Bob', 'Goodman')""")

# inserts new author's ISBN
cursor = cursor.execute("""INSERT INTO author_ISBN (id, isbn) VALUES ('6', '1234567899')""")

# inserts new author's book
cursor = cursor.execute("""INSERT INTO titles (isbn, title, edition, copyright) VALUES ('1234567899', 'Goodman Code', '99', '1999')""")

print()

# displays the titles of books including the new author's book
print(pd.read_sql('SELECT * FROM titles ORDER BY title ASC', connection))

print()

# displays the authors of books including the new author
print(pd.read_sql('SELECT * FROM authors ORDER BY last DESC', connection))

print()

# displays the ISBNs including the new author's ISBN
print(pd.read_sql('SELECT * FROM author_ISBN ORDER BY id DESC', connection))


```
